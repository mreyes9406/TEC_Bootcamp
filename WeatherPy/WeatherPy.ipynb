### Analysis
* There is a clear relationship between the latitude and the temperature: as the latitude's absolute value increases, the temperature drops, which is consistent with the fact that the temperature is lower as we approach to the poles. It can also be told that there are much more records of cities in the northern than in the southern hemisphere. This might be due to the fact that there is more continental mass in that hemishphere, thus the number of cities will also be higher.  

* There seems not to be a relationship between the cloudiness and the latitude. Despite this, two data clusters can be observed: one of cloudy cities (around 90% cloudiness) and the other of nearly null cloudiness. Both clusters are located in the northern hemisphere.

* For the current day, the speed of the wind is under 15 MPH for more than 80% of the cities, regardless of their latitude. Still there are a few cities with speeds significantly higher than 15 MPH; the majority of these cities are located in the northern hemisphere (positive lattitudes). According to this information, the highest speeds only occur in the northernmost cities of the planet. This might not be true due to the fact that vast majority of cities lies in that hemishphere, so the number of records for the southernmost cities in the planet is comparably small, so there is not enough information.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Show graphs
%matplotlib inline
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API calls

```python
# Create lists to store retrieved values and flag to count iterations
names = []
countries = []
longs = []
lats = []
dates = []
temps_max = []
hum_pcts = []
cloud_pcts = []
wind_sps = []
flag = 1

# Print header
print("Beginning Data Retrieval")
print("------------------------------------------")

# Perform successive API calls for cities 
for city in cities:
    response = requests.get(f"http://api.openweathermap.org/data/2.5/weather?q={city}&units=imperial&appid={api_key}")
    response_json = response.json()
    
    # Store values in corresponding lists for successful API responses
    if int(response.status_code) == 200:
                 
        # Variables to store specific information taken from JSON file
        name = str(response_json["name"])
        country = str(response_json["sys"]["country"])
        lon = float(response_json["coord"]["lon"])
        lat = float(response_json["coord"]["lat"])
        date = int(response_json["dt"])
        temp_max = float(response_json["main"]["temp_max"])
        hum_pct = float(response_json["main"]["humidity"])
        cloud_pct = float(response_json["clouds"]["all"])
        wind_sp = float(response_json["wind"]["speed"])
        
        # Append values to corresponding lists
        names.append(name)
        countries.append(country)
        longs.append(lon)
        lats.append(lat)
        dates.append(date)
        temps_max.append(temp_max)
        hum_pcts.append(hum_pct)
        cloud_pcts.append(cloud_pct)
        wind_sps.append(wind_sp)
        
        # Print log
        print(f"Processing Record {flag} | {name}")
        flag += 1
    
    # Send error message and skip unfound cities
    else:
        print("City not found. Skipping...")
        flag += 1

# Print footer
print("------------------------------------------")
print("Finished Data Retrieval")
```

```python
# Create dataframe from lists constructed with API responses
df = pd.DataFrame({
    "City": names,
    "Country": countries,
    "Latitude": lats,
    "Longitude": longs,
    "Date": dates,
    "Max Temperature": temps_max,
    "Humidity Percent": hum_pcts,
    "Cloudiness": cloud_pcts,
    "Wind Speed": wind_sps
    
})

# Save and display dataframe
df.to_csv(output_data_file)
df.head()
```

#### Temperature vs. Latitude Plot

```python
plt.scatter(df["Latitude"], df["Max Temperature"], marker="o", edgecolors="black", alpha=0.8)
plt.grid()
plt.title(f"Max Temperature vs. City Latitude ({datetime.date.today()})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (°F)")
plt.savefig("temp_vs_lat.png")
plt.show()
```
